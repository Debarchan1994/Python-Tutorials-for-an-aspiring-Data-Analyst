Example 2: Check if the 2 numbers are same and divisible by 2

```python
num_1 = input('Enter the 1st number')
num_2 = input('Enter the 2nd number')
num_1 = int(num_1)
num_2 = int(num_2)
```

```python
'''The % operator used in this example is called the modulo operator and it returns the remainder of the division of one number 
by another'''
```

```python
if num_1 == num_2 and num_1%2 == 0 and num_2%2 == 0:
    print("Congratulations !! both numbers are same and divisible by 2")

elif num_1 == num_2 and num_1%2 != 0 and num_2%2 != 0:
    print("Both the numbers are same but not divisible by 2")
    
else:
    print("Numbers are not same")
```

```python
'''Same example with nested if statement'''

if num_1 == num_2: 
    if num_1%2 == 0 and num_2%2 == 0:
        print("Congratulations !! both numbers are same and divisible by 2")
    
    elif num_1%2 != 0 and num_2%2 != 0:
        print("Both the numbers are same but not divisible by 2")
    
else:
    print("Numbers are not same")
```

# What are loops?

Loops are used to iterate over sequences and perform some repetitive task.

There are mainly 2 types to loops that are used in python for loop and while loop.

There is another kind of loop that is do-while loop but this is not used frequently.

For loop: 
    
    It is generally used to  iterate over a sequence (such as a list, tuple, string, or range) or other iterable objects.
    
Examples are below

```python
my_list = [10,20,30,40,50,90,200]

for num in my_list:
    print(num)
```

```python
'''In this example range keyword is used it is used to define a sequence of numbers. 
It is commonly used in for loops to iterate over a sequence of numbers. 

3 commonly used syntaxes for range are :

1) range(stop) : Generates a sequence of numbers from 0 to (stop - 1)
2) range(start, stop) : Generates a sequence of numbers from start to (stop - 1)
3) range(start , stop, step_size): Generates a sequence of numbers from start to (stop - 1) with a step size of step size'''

for num in range(0,10):
    print(num)
```

```python
word = 'Anaconda'

for char in word:
    print(char)
```

While loop: 
    
    The while loop is used to repeatedly execute a piece of code as long as the specified 
    condition is satisfied.
    
    
    Below are some of the examples of while loop.

```python
count = 0
while count < 10:
    print(count)
    count = count + 1
```

```python
'''This is an example of infinite loop'''

while True:
    print('*')
```

```python
'''Infinite While loop is used in many cases to keep interating till a termination condition is met.
This allows us to input multiple entries without requering to execute the code every time.
Think about then you type you password wrong it allows you to retype it again and again 
and after some failed attempt the program terminates'. This can be a typical example of While loop'''

password = "Python@1234" ## think this is stored in the database and you cannot see it but needs to 
                         ## be matched in order to grant you access

while True:
    user_input = input("Enter your password: ")
    if user_input == password:
        print("Access granted")
        break
    else:
        print("Incorrect password. Try again.")
```

Example 1 : Count the number of elements starting with 'a' and number of elements containing atleast one 'a' in the list given below using while loop

```python
my_list = ['suman', 'ana', 'veronica', 'philipp', 'justin', 'anthony', 'ajay']
```

```python
count = 0
starting_a = 0
containing_a = 0

'''In the below while loop the condition is set as count <= len(my_list) -1. Here 1 is substracted
from the lenght of the list to avoid index out of bound error. We know that index of a list
starts from 0 and the length of the list  is 7 in this case. Now if we don't substract 1 then the
loop will run till the value of the count variable is greater than 7, in that case the below if 
statements will throw error during the last iteration as there is no element in the loop with
index 7 as the indexes are from 0 till 6.'''

while count <= len(my_list) - 1:
    
    if list(my_list[count])[0] == 'a': ## list of a string 'ana' will give ['a','n','a']. 
        starting_a = starting_a + 1
        
    if 'a' in list(my_list[count]):
        containing_a = containing_a + 1
        
    count = count + 1
    
print('number of elements starting with a is/are: ', starting_a)
print('number of elements containing atleast one a is/are: ', containing_a)
```

Practice Example 2: In the below list calculate the sum of the elements untill the last '11' is encountered using only while loop and if statement wherever necessary.
For example, here the last 11 is before the element 61 so the sum will be calculated till 100.

```python
a = [1,2,4,6,10,23,11,34,11, 100, 11, 61, 50]
```

```python
i = 0
sum_ = 0
number_11 = []

while i <= len(a) -1:
    
    if a[i] == 11:
        number_11.append(i)
    
    i = i + 1

i = 0 

while i < number_11[-1]:
    
    sum_ = sum_ + a[i]
    i = i + 1
    
print('sum is :', sum_)
```

Using loops on top of dictionaries

```python
a = {'age': 29, 'salary': 5000, 'jobs': 2}
```

```python
for i,j in a.items():  ##In case of dictionary we use dictionary.item() syntax as it catches both the key and its value
    print(i,j)
```

```python
for i in a:     ## If in case we use normal loop syntax like incase of list or tuple then only the keys are catched in each 
    print(i)    ## iteration and not the values
        
```

```python
for i in a.keys(): ## dictionary.keys() catches on the keys  and dictionary.values() catches only the values
    print(i)

for i in a.values():
    print(i)
```
